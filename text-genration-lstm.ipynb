```python
import tensorflow as tf
from tensorflow .keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([train_content[:3000]])
```

```python
#converting supervised for nxt tect genartion
```

```python
input_sequences = []
for sentences in train_content[:3000].split("/n"):
  #print(sentences)
  tokenize_sentence = tokenizer.texts_to_sequences([sentences])[0]

  for i in range(1,len(tokenize_sentence)):
    input_sequences.append(tokenize_sentence[:i+1])
```

```python
input_sequences
```

```python
# making all senetences in same shape
# zero padding
```

```python
max_len = max([len(x) for x in input_sequences])
```

```python
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences , maxlen = 338 ,padding = "pre")
```

```python
padded_input_sequences
```

```python
padded_input_sequences[:,:-1]
```

```python
x = padded_input_sequences[:,:-1]
y = padded_input_sequences[:,-1]
```

```python
y
```

```python
# making multi class
#using one hot encoding
#[010000] [100000 ]
```

```python
x.shape,y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes = len(tokenizer.word_index)+1)
```

```python
y.shape,y
```

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Embedding, Flatten ,Dropout ,Embedding,LSTM
```

```python
model = Sequential()
model.add(Embedding(len(tokenizer.word_index)+1,100,input_length = 338 ))
model.add(LSTM(300))
#model.add(LSTM(300))
model.add(Dense(len(tokenizer.word_index)+1,activation = "softmax"))

model.build(input_shape=(None, max_len-1))
```

```python
model.compile(loss = "categorical_crossentropy",optimizer = "adam",
              metrics = ["accuracy"])
```

```python
model.summary()
```

```python
model.fit(x,y,epochs = 10,validation_split = 0.2)
```
